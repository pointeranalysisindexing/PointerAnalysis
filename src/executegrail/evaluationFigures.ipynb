```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("GrailQueriesPerQuery.csv")
df = pd.DataFrame(data, columns=['Benchmark', 'No of queries', 'Grail'])

df.set_index('No of queries', inplace=True)
df.groupby('Benchmark')['Grail'].plot(legend=True, ylabel='Average runtime per query (ms)', xlabel='# queries')
```
